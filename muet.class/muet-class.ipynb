```python
# Student data - list of dictionaries
# Each student has a name, grade, and age
students = [
    {'name': 'Ali',  'grade': 85, 'age': 20},
    {'name': 'Sara', 'grade': 92, 'age': 22},
    {'name': 'John', 'grade': 78, 'age': 19},
    {'name': 'Lena', 'grade': 88, 'age': 21},
]
```

```python
# Task 1: Sort by grade - lowest grade first (ascending order)
sorted_by_grade_low = sorted(students, key=lambda s: s['grade'])
print('Task 1 - Grade Low to High:')
for s in sorted_by_grade_low:
    print(s)
```

```python
# Task 2: Sort by grade - highest grade first (descending order)
# reverse=True flips the order from high to low
sorted_by_grade_high = sorted(students, key=lambda s: s['grade'], reverse=True)
print('Task 2 - Grade High to Low:')
for s in sorted_by_grade_high:
    print(s)
```

```python
# Task 3: Sort by age - youngest student first (ascending order)
sorted_by_age = sorted(students, key=lambda s: s['age'])
print('Task 3 - Youngest First:')
for s in sorted_by_age:
    print(s)
```

```python
# Task 4: Sort by name alphabetically (A to Z)
sorted_by_name = sorted(students, key=lambda s: s['name'])
print('Task 4 - Name A to Z:')
for s in sorted_by_name:
    print(s)
```

```python
# Task 5: Sort by highest grade first
# If two students have the same grade, sort them by name (A to Z)
# Using a tuple in lambda: first sort by grade (descending), then by name (ascending)
sorted_by_grade_then_name = sorted(students, key=lambda s: (-s['grade'], s['name']))
print('Task 5 - Highest Grade First, then Name A-Z if same grade:')
for s in sorted_by_grade_then_name:
    print(s)
```
