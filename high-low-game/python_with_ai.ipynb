```python
# Real-world: a receipt generator
def print_receipt(customer, *items):
    print(f'\n--- Receipt for {customer} ---')
    total = 0
    for name, price in items:
        print(f'  {name:<15} Rs. {price}')
        total += price
    print(f'  {"TOTAL":<15} Rs. {total}')
print_receipt('Ali',
    ('Burger', 350),
    ('Fries', 150),
    ('Cola', 100))


a = [1,2,3,4]
b =  ((),(1,2,4))
```

```python
# Basic **kwargs
def create_profile(**details):
    print(type(details))   # <class 'dict'>
    for key, value in details.items():
        print(f'  {key}: {value}')

create_profile(name='Ali', age=20, city='Karachi')
create_profile(name='Sara', role='Admin', active=True, level=5)
```

```python
    # Real-world: HTML tag builder
def build_tag(tag, content, **attributes):
    
    attrs = ' '.join(f'{k}="{v}"' for k, v in attributes.items())
    
    return f'<{tag} {attrs}>{content}</{tag}>'

print(build_tag('a', 'Click here', href='https://muet.edu.pk', target='_blank'))

# <a href="https://muet.edu.pk" target="_blank">Click here</a>

print(build_tag('img', '', src='photo.jpg', width='200', alt='Profile'))
```

```python
# Order MUST be: regular params → *args → **kwargs
def send_email(sender, *recipients, **options):
    print(f'From: {sender}')
    print(f'To: {', '.join(recipients)}')
    subject = options.get('subject', '(No Subject)')
    priority = options.get('priority', 'normal')
    print(f'Subject: {subject} | Priority: {priority}')

send_email('admin@muet.edu.pk',
           'ali@gmail.com', 'sara@gmail.com', 'john@gmail.com',
           subject='Assignment Reminder', priority='high')
```

```python
#Section 2: map() — Transform Every Element
# map(function, iterable) → applies function to each element

prices = [100, 250, 80, 500, 320]

# Apply 17% tax to all prices
with_tax = list(map(lambda p: round(p * 1.17, 2), prices))
print(with_tax)  # [117.0, 292.5, 93.6, 585.0, 374.4]

#def add_tax(p):
#    return round(p * 1.17, 2)

# lambda p: round(p * 1.17, 2)
# Convert Celsius temps to Fahrenheit
celsius = [0, 20, 37, 100]
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print(fahrenheit)  # [32.0, 68.0, 98.6, 212.0]

# Clean and capitalize a list of names
raw_names = ['  ali  ', 'SARA', 'john  ', '  MIKE']
clean_names = list(map(lambda n: n.strip().title(), raw_names))
print(clean_names)  # ['Ali', 'Sara', 'John', 'Mike']
```
